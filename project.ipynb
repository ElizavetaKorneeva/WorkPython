```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
from sklearn.cluster import KMeans

%matplotlib inline 
```

```python
data = pd.read_excel('Python_data.xlsx')
data.head()
```

```python
data_for_clust = data.drop(columns=['Regions', 'min_rate', 'max_rate', 'mean_rate']).values
data_plot = data.drop(columns=['Regions', 'min_rate', 'max_rate']).values

data_for_clust[0]
```

```python
from sklearn import preprocessing
data_norm = preprocessing.scale(data_for_clust)
plot_norm = preprocessing.scale(new)
```

```python
km = KMeans(n_clusters=6).fit(data_norm)
```

```python
km.labels_ +1
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(data_plot[:,0], data_plot[:,1], c=km.labels_)
plt.show()
```

```python
dataK=data
dataK['group_no'] = km.labels_ + 1

writer = pd.ExcelWriter('Phyton_data.xlsx')
dataK.to_excel(writer,'KMeans')
writer.save()
```
